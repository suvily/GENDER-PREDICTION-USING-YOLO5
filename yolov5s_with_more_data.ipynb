## Extract face from YOLO

```python
from PIL import Image, ImageDraw, ImageFont
import torch, time, os
from google.colab import drive
from IPython.display import clear_output
```

```python
def extractFace(IMG, FaceDetector, threshold=0.50):
    extractedFaces = []
    FaceDetections = FaceDetector(IMG).pandas().xyxy[0] #facedetector is the genederModel
    for detection in FaceDetections.values:
        xmin, ymin, xmax, ymax, confidence = detection[:5]
        if confidence >= threshold:
            bb = [(xmin, ymin), (xmax, ymax)]
            w, h = xmax - xmin, ymax - ymin
            currentFace = IMG.crop((xmin, ymin, w+xmin, h+ymin))
            extractedFaces.append(currentFace)

    return extractedFaces
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('results.csv')
df.columns
```

```python
df.shape
```

```python
# Generate a sequence of integers to represent the epoch numbers
epochs = range(1, 16)
 
# Plot and label the training and validation loss values
plt.plot(epochs, df['      train/box_loss'], label='Training box Loss')
plt.plot(epochs, df['        val/box_loss'], label='Validation box Loss')
 
# Add in a title and axes labels
plt.title('Training and Validation Box Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
 
# Set the tick locations
plt.xticks(np.arange(0, 16, 2))
 
# Display the plot
plt.legend(loc='best')
plt.show()
```

```python
# Generate a sequence of integers to represent the epoch numbers
epochs = range(1, 16)
 
# Plot and label the training and validation loss values
plt.plot(epochs, df['      train/obj_loss'], label='Training Obj Loss')
plt.plot(epochs, df['        val/obj_loss'], label='Validation Obj Loss')
 
# Add in a title and axes labels
plt.title('Training and Validation Object Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
 
# Set the tick locations
plt.xticks(np.arange(0, 16, 2))
 
# Display the plot
plt.legend(loc='best')
plt.show()
```

```python
# Generate a sequence of integers to represent the epoch numbers
epochs = range(1, 16)
 
# Plot and label the training and validation loss values
plt.plot(epochs, df['      train/cls_loss'], label='Training Class Loss')
plt.plot(epochs, df['        val/cls_loss'], label='Validation Class  Loss')
 
# Add in a title and axes labels
plt.title('Training and Validation Class Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
 
# Set the tick locations
plt.xticks(np.arange(0, 16, 2))
 
# Display the plot
plt.legend(loc='best')
plt.show()
```
